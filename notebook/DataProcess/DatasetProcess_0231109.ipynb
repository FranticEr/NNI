```python
import matplotlib.pyplot as plt

def plot_fun(func):
    plt.figure()  # 每次调用函数创建一个新的子图
    plt.xlabel('i')
    plt.ylabel('fun(i)')
    plt.title('fun(i) Output Curve')
    
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        plt.plot(result)
        return result
    
    return wrapper

@plot_fun
def fun(x):
    return x

for i in range(100):
    fun(i)

plt.show()  # 显示所有子图
```
