```python
from autogluon.tabular import TabularDataset,TabularPredictor
import pandas as pd
train_data_train_path=r"D:\project_meta\NNproject\NNI\output\bandpoweruniform.csv"
data=pd.read_csv(train_data_train_path,encoding="gbk")
from sklearn.model_selection import train_test_split
#train_data,test_data=train_test_split(train_data,test_size=0.25)
train_data,test_data=train_test_split(data,test_size=0.25)
```

```python
label="LEVEL"
metric = 'roc_auc'  # specify your evaluation metric here
metric = 'roc_auc'
predictor = TabularPredictor(label).fit(train_data.drop(columns=['KSS','ID']), time_limit=600, presets='best_quality')
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,f1_score,classification_report
predictor = TabularPredictor.load(r"D:\project_meta\NNproject\NNI\notebook\AutogluonModels\ag-20231111_063512")
pred=predictor.predict(test_data)
#F1
cy=confusion_matrix(pred, test_data[label])
print("confusion_matrix:",cy)
dc=accuracy_score(pred, test_data[label])
print("accuracy_score",dc)
recall=recall_score(pred, test_data[label],average='micro')
print("recall_score",recall)
f1=f1_score(pred, test_data[label],average='micro')
print("f1_score",f1)
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
total_correct = np.sum(np.diag(cy))
total_samples = np.sum(cy)

accuracy = total_correct / total_samples
print("Overall Accuracy:", accuracy)
# 
plt.figure(figsize=(8, 6))  # 
sns.heatmap(cy, annot=True, fmt="d", cmap="Blues")  # 
plt.title("Confusion Matrix")  # 
plt.xlabel("Predicted Label")  # x
plt.ylabel("True Label")  # y
plt.show()  # 
```
